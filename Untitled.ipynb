```python
import cv2
import os

# Define the paths for the input and output directories
input_dir = 'images'  # Directory where the input images are located
output_dir = 'cropped images'  # Directory to save the cropped images

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Coordinates and dimensions for cropping (adjust these values as needed)
crop_x = 75
crop_y = 150
crop_width = 1050  # Width of the cropped region
crop_height = 1000  # Height of the cropped region

# Get a list of all files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')]

# Sort filenames numerically
input_files.sort(key=lambda x: int(''.join(filter(str.isdigit, x)) or 0))

# Ensure there are at least 14 images to process
if len(input_files) < 14:
    raise ValueError("Not enough images in the input directory. At least 14 images are required.")

# Process only the first 14 images
for i in range(14):
    filename = input_files[i]
    
    # Full path to the input image
    input_image_path = os.path.join(input_dir, filename)
    
    # Read the input image
    image = cv2.imread(input_image_path)
    
    # Check if the image was loaded successfully
    if image is not None:
        # Crop the image using the defined coordinates and dimensions
        cropped_image = image[crop_y:crop_y + crop_height, crop_x:crop_x + crop_width]
        
        # Define the output image filename
        output_image_filename = f'cropped_{i+1:02d}.jpg'  # e.g., cropped_01.jpg, cropped_02.jpg, ...
        
        # Full path to the output image
        output_image_path = os.path.join(output_dir, output_image_filename)
        
        # Save the cropped image
        cv2.imwrite(output_image_path, cropped_image)
        print(f'Cropped image saved: {output_image_path}')
    else:
        print(f'Failed to load image: {input_image_path}')

print('Cropping completed.')
```

```python
import cv2
import numpy as np
import os

# Define the paths for the input and output directories
input_dir = 'gray scale'  # Directory where the input images are located
output_dir = 'log_transformed_images'  # Directory to save the transformed images

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get a list of all files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')]

# Ensure there are images to process
if len(input_files) == 0:
    raise ValueError("No images found in the input directory.")

# Process each image in the directory
for filename in input_files:
    # Full path to the input image
    input_image_path = os.path.join(input_dir, filename)
    
    # Read the input image
    image = cv2.imread(input_image_path)
    
    # Check if the image was loaded successfully
    if image is not None:
        # Convert the image to float32 for precision in calculations
        image_float = np.float32(image)
        
        # Apply log transformation
        c = 255 / np.log(1 + np.max(image_float))
        log_transformed = c * np.log(1 + image_float)
        
        # Convert back to uint8
        log_transformed = np.uint8(log_transformed)
        
        # Define the output image filename
        output_image_filename = f'log_transformed_{filename}'
        
        # Full path to the output image
        output_image_path = os.path.join(output_dir, output_image_filename)
        
        # Save the log-transformed image
        cv2.imwrite(output_image_path, log_transformed)
        print(f'Log-transformed image saved: {output_image_path}')
    else:
        print(f'Failed to load image: {input_image_path}')

print('Log transformation completed for all images.')
```

```python
import cv2
import os
import numpy as np

# Define paths
input_folder = 'median filter'  # Replace with the path to your input folder
output_folder = 'canny edge detection'  # Replace with the path to your output folder

# Create the output folder if it does not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Set Canny edge detection thresholds
low_threshold = 10
high_threshold = 140

# Process each image in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an image
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
        # Load the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        
        # Apply Gaussian Blur
        blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
        
        # Perform Canny edge detection
        edges = cv2.Canny(blurred_image, low_threshold, high_threshold)
        
        # Save the edge-detected image
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, edges)
        
        print(f'Processed and saved: {filename}')

print('Edge detection completed for all images.')
```

```python
import cv2
import numpy as np
import os

# Define paths
input_folder = 'canny edge detection'  # Replace with the path to your binary segmented images
output_folder = 'diluted images'  # Replace with the path to save processed images

# Create the output folder if it does not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Define the structuring element for dilation
kernel_size = 5  # Size of the structuring element (kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (kernel_size, kernel_size))

# Process each image in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an image
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
        # Load the image (assuming the image is already binary/segmented)
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        
        # Threshold the image to binary (if needed)
        _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

        # Apply dilation
        dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
        
        # Save the processed image
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, dilated_image)
        
        print(f'Processed and saved: {filename}')

print('Dilation completed for all images.')
```

```python
import cv2
import os

# Define the paths for the input and output directories
input_dir = 'diluted images'  # Directory where the input images are located
output_dir = 'adaptive_thresholded_images'  # Directory to save the thresholded images

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get a list of all files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')]

# Ensure there are images to process
if len(input_files) == 0:
    raise ValueError("No images found in the input directory.")

# Process each image in the directory
for filename in input_files:
    # Full path to the input image
    input_image_path = os.path.join(input_dir, filename)
    
    # Read the input image
    image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if the image was loaded successfully
    if image is not None:
        # Apply Gaussian blur to the image (optional but often improves results)
        blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
        
        # Apply adaptive thresholding
        adaptive_thresholded = cv2.adaptiveThreshold(
            blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
        )
        
        # Define the output image filename
        output_image_filename = f'adaptive_thresholded_{filename}'
        
        # Full path to the output image
        output_image_path = os.path.join(output_dir, output_image_filename)
        
        # Save the thresholded image
        cv2.imwrite(output_image_path, adaptive_thresholded)
        print(f'Adaptive-thresholded image saved: {output_image_path}')
    else:
        print(f'Failed to load image: {input_image_path}')

print('Adaptive thresholding completed for all images.')
```

```python
import cv2
import numpy as np
import os

# Define paths
input_folder = 'adaptive_thresholded_images'  # Replace with the path to your images
output_folder = 'flood fill'  # Replace with the path to save processed images

# Create the output folder if it does not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Parameters for flood fill
seed_point = (200, 800)  # Example seed point
new_val = 200  # Fill color (e.g., white)
lo_diff = 10  # Lower brightness/color difference
up_diff = 10  # Upper brightness/color difference

# Process each image in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an image
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
        # Load the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        
        # Make a copy of the image to preserve the original
        flood_filled_image = image.copy()
        
        # Apply flood fill
        mask = np.zeros((image.shape[0] + 2, image.shape[1] + 2), np.uint8)  # Mask needs to be 2 pixels larger than image
        cv2.floodFill(flood_filled_image, mask, seed_point, new_val, lo_diff, up_diff, flags=cv2.FLOODFILL_FIXED_RANGE)
        
        # Save the processed image
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, flood_filled_image)
        
        print(f'Processed and saved: {filename}')

print('Flood fill completed for all images.')
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Define paths
input_folder = 'flood fill'  # Replace with the path to your flood filled images

# Initialize a list to store the areas
areas = []

# Process each image in the input folder
for filename in sorted(os.listdir(input_folder)):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
        # Load the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        
        # Calculate the area of the fungal growth by counting zero (black) pixels
        area = image.size - cv2.countNonZero(image)
        areas.append(area)

        print(f'Processed {filename}: Area = {area}')

# Generate a time array based on the number of images
time = list(range(1, len(areas) + 1))  # Assuming each image represents a day

# Plot the growth rate
plt.figure(figsize=(10, 6))
plt.plot(time, areas, marker='o', linestyle='-', color='b')
plt.title('Fungal Growth Rate Over Time')
plt.xlabel('Time (Days)')
plt.ylabel('Fungal Growth Area (Pixels)')
plt.grid(True)
plt.savefig('fungal_growth_rate.png')  # Save the plot as an image
plt.show()
```
